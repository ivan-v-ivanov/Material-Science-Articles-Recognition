```python
import requests
from lxml import html
```

```python
url = 'https://www.nature.com'

params = {'searchType':'journalSearch',
          'sort': 'PubDate',
          'type': 'article',
          'page':'1'}

headers = {'User-Agent':'Mozilla/5.0 (X11; Linux i686; rv:102.0) Gecko/20100101 Firefox/102.0',
          }
```

```python
articles_database = {'title':[],
                    'publication_date':[],
                    'abstract':[]}
while True:
    response = requests.get(url+'/nmat/research-articles', 
                            params = params, headers=headers)

    if response.ok:
        print(f"Page #{params['page']} | Response: OK\n")
        dom = html.fromstring(response.text)
    else:
        print(response)
        pass
        
    articles = dom.xpath("//h3[contains(@class, 'c-card__title')]")
    next_button = dom.xpath("//li[contains(@data-page, 'next')]//@href")
    
    for i, article in enumerate(articles, 1):
        href = article.xpath(".//a[contains(@class, 'c-card__link u-link-inherit')]/@href")[0]
    
        article_response = requests.get(url+href, headers=headers)
        
        if article_response.ok:
            article_dom = html.fromstring(article_response.text)
            title = article_dom.xpath("//h1/text()")[0]
            publication_date = article_dom.xpath("//a[contains(@data-track-action, 'publication date')]/time/@datetime")[0]
            abstract = ''.join(article_dom.xpath("//div[contains(@id, 'Abs1-content')]/descendant::*/text()"))
            
            print(f'\t{i}. {title}\n\t{publication_date}')
            #print(f'{abstract}\n')
            articles_database['title'].append(title)
            articles_database['publication_date'].append(publication_date)
            articles_database['abstract'].append(abstract)
            
        else:
            print(article_response)
            pass    

    if not next_button:
        break
        
    params['page'] = str(int(params['page'])+1)
```

```python
import pandas as pd
```

```python
database = pd.DataFrame.from_dict(articles_database)
```

```python
database.to_csv('nature_materials.csv')
```
